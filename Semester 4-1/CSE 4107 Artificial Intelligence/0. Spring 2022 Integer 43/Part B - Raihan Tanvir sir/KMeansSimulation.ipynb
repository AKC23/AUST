```python
df1 = df.copy()

# centroids = df.sample(2).reset_index(drop=True)
# centroids.reset_index(drop=True).to_numpy()
centroids = df.loc[[0, 3], :].reset_index(drop=True)

df1['dist_centroid_1'] = np.linalg.norm(df - centroids.loc[0], axis=1)
df1['dist_centroid_2'] = np.linalg.norm(df - centroids.loc[1], axis=1)
print(df1)

df1['Cluster'] = np.where(df1.dist_centroid_1 <= df1.dist_centroid_2, 1, 2)
print(df1)

df1['PrevCluster'] = np.zeros(len(df))
print(df1)
```

```python
movement = True

while (movement):
    centroids = df1.groupby('Cluster')['X1', 'X2'].mean()
    centroids = centroids.reset_index(drop=True)
    print("Centroids:\n", centroids)

    df1['dist_centroid_1'] = np.linalg.norm(df - centroids.loc[0], axis=1)
    df1['dist_centroid_2'] = np.linalg.norm(df - centroids.loc[1], axis=1)
    df1['PrevCluster'] = df1['Cluster']
    df1['Cluster'] = np.where(df1.dist_centroid_1 <= df1.dist_centroid_2, 1, 2)
    print(df1)
    
    movement = (df1['PrevCluster'] != df1['Cluster']).any()
    
    if (movement):
        print('Data Moved\n')
    else:
        print('No Movement of Data\n')
        
    count = count + 1
```
